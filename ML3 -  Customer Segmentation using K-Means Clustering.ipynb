# **now out of all attribute we'll choose 2 columns for clustering i.e Annual Income (k$)	Spending Score (1-100)**

```python
X = dataset.iloc[:,[3,4]].values                  #indexing starts from 0
```

```python
print(X)
```

# now we have to decide number of clusters to be made

**for that we'll use**

WCSS --> [Within cluster sum of square]

```python
# finding the WCSS

wcss = []

for i in range(1,11):
  kmeans = KMeans(n_clusters=i , init='k-means++' , random_state=10)
  kmeans.fit(X)

  wcss.append(kmeans.inertia_)
```

```python
sns.set()
plt.plot(range(1,11), wcss)
plt.title('The Elbow Point Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
